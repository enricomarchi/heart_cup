```python
import pandas as pd
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import math
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
orari = pd.date_range(start="09:00", end="20:00", freq="5min").strftime('%H:%M')
tatami = ['Tatami ' + str(i) for i in range(1,7)]
timetable_sabato = pd.DataFrame(index=orari, columns=tatami)
timetable_domenica = pd.DataFrame(index=orari, columns=tatami)
timetable_sabato = pd.DataFrame(index=orari, columns=tatami).fillna('')
timetable_domenica = pd.DataFrame(index=orari, columns=tatami).fillna('')
timetable_sabato['Disponibile'] = True
timetable_domenica['Disponibile'] = True
```

```python
iscrizioni = pd.read_parquet("iscrizioni.parquet")
iscrizioni.info()
```

```python
atleti = iscrizioni.drop_duplicates(subset=['NOME', 'COGNOME']).loc[:, ["COGNOME", "NOME", "DATA DI NASCITA (AAAA-MM-GG)", "SESSO", "Kyu/Kup", "Dan", "PESO / KG", "ALTEZZA / CM"]].sort_values(by=['COGNOME', 'NOME'], ascending=True)
atleti.reset_index(drop=True, inplace=True)
atleti["DATA DI NASCITA (AAAA-MM-GG)"] = pd.to_datetime(atleti["DATA DI NASCITA (AAAA-MM-GG)"]).dt.date
atleti["SESSO"] = atleti["SESSO"].str.upper()
atleti["Kyu/Kup"] = atleti["Kyu/Kup"].astype(int)
atleti["Dan"] = atleti["Dan"].astype(int)
atleti["PESO / KG"] = atleti["PESO / KG"].astype(int)
atleti["ALTEZZA / CM"] = atleti["ALTEZZA / CM"].astype(int)
atleti
```

```python
categorie = iscrizioni["CATEGORIA"].unique()
iscritti_per_categoria = iscrizioni.groupby('CATEGORIA')['NOME'].count()
```

```python
iscritti_colorate = iscritti_per_categoria[iscritti_per_categoria.index.str.contains('YELLOW|ORANGE|GREEN|BLUE')].sum()
iscritti_colorate
```

```python
iscritti_marroni = iscritti_per_categoria[iscritti_per_categoria.index.str.contains('BROWN')].sum()
iscritti_marroni
```

```python
iscritti_nere = iscritti_per_categoria[iscritti_per_categoria.index.str.contains('BLACK')].sum()
iscritti_nere
```

```python
import math

# Variabili di tempo
tempo_kata = 1  # 2 minuti e mezzo per gli incontri di KATA
tempo_a_punteggio = 2.5  # 4 minuti fino agli ottavi di finale in KUMITE
tempo_a_bandierine = 2  # 2 minuti e mezzo dagli ottavi alla finale in KUMITE

# Funzione per calcolare il tempo totale per ciascuna categoria
def calcola_tempo(categoria, conteggio_iscritti):
    if "TEAM" in categoria:
        return 0
    elif "KATA" in categoria:
        return math.ceil(conteggio_iscritti * tempo_kata)
    elif "KUMITE" in categoria:
        round_ = conteggio_iscritti
        tempo_totale = 0
        while round_ > 1:
            incontri = round_ // 2

            # Fase a punteggio (fino agli ottavi)
            if round_ > 16:
                tempo_round = tempo_a_punteggio
            # Fase a bandierine (dagli ottavi alla finale)
            else:
                tempo_round = tempo_a_bandierine

            tempo_totale += incontri * tempo_round
            round_ = incontri + (round_ % 2)
        return math.ceil(tempo_totale)
    else:
        return 0

# Applica la funzione per calcolare il tempo totale
tempi_per_categoria = iscritti_per_categoria
tempi_per_categoria = iscritti_per_categoria.reset_index()
tempi_per_categoria['TEMPO (MINUTI)'] = tempi_per_categoria.apply(lambda row: calcola_tempo(row['CATEGORIA'], row['NOME']), axis=1)
tempi_per_categoria.rename(columns={'NOME': 'ATLETI'}, inplace=True)
tempi_per_categoria['Tipo'] = tempi_per_categoria['CATEGORIA'].apply(lambda x: 'KATA' if 'KATA' in x else ('KUMITE' if 'KUMITE' in x else 'ALTRO'))
tempi_per_categoria
```

```python
atleti_per_categoria = iscrizioni.groupby('CATEGORIA').apply(
    lambda x: (x.index.astype(str) + "-" + x['COGNOME'] + " " + x['NOME']).drop_duplicates().values.tolist()
)
atleti_per_categoria
```

```python

# Funzione per assegnare la categoria
def assegna_categoria(timetable, categoria, minuti):
    migliori_tatami = []
    
    for ora in timetable.index:
        ora_inizio = datetime.strptime(ora, '%H:%M')
        ora_fine = (ora_inizio + timedelta(minutes=minuti)).strftime('%H:%M')
        
        if ora_fine > timetable.index[-1]:
            continue  # Non c'è abbastanza tempo nella giornata

        tatami_candidati = []
        
        for tatami in timetable.columns:
            if timetable.loc[ora:ora_fine, tatami].eq('').all():
                tatami_candidati.append(tatami)
                
        if len(tatami_candidati) >= 1:
            migliori_tatami = tatami_candidati[:2]  # Prendi al massimo due tatami
            break
    
    if migliori_tatami:
        for tatami in migliori_tatami:
            timetable.loc[ora:ora_fine, tatami] = categoria
        return True

    return False  # Non è stato possibile assegnare la categoria

# Assegnazione delle categorie ai tatami
for _, row in tempi_per_categoria.iterrows():
    categoria = row['CATEGORIA']
    minuti = row['TEMPO (MINUTI)']
    
    assegnato = assegna_categoria(timetable_sabato, categoria, minuti)
    if not assegnato:
        assegnato = assegna_categoria(timetable_domenica, categoria, minuti)

    if not assegnato:
        print(f"Non è stato possibile assegnare la categoria {categoria}")

# Aggiungi le pause pranzo
pause_pranzo = pd.date_range(start="12:00", end="13:30", freq="15min").strftime('%H:%M')
for tatami in tatami:
    timetable_sabato.loc[pause_pranzo, tatami] = 'Pausa Pranzo'
    timetable_domenica.loc[pause_pranzo, tatami] = 'Pausa Pranzo'
```
